### 1: high pass -> notch filter
    - notch targets narrow frequency band VS high pass removes a broad range of low frequences
    **highpass** 
    removes baseline drifts (slow  unwanted changes in signals baseline level) removes 0.5/1 Hz etc
    **notch fitler**
    - remove power line noise( 50 or 60 Hz) from  signal, creating a cleaner data set more suitable for extracting meaningful features related to emotional states
    - improves the accuracy of emotion classification algorithms. 
    notch filtering = initial step 
    
### 2: Band Pass filtering
    - filter to extract specific frequencies by iodentifying the low cutoff and high cutoff frequencis 

```python
import numpy as np
from scipy.signal import butter, lfilter
import matplotlib.pyplot as plt

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order)
    y = lfilter(b, a, data)
    return y
    
## HIGHPASS FILTER CODE
def butter_highpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

def butter_highpass_filter(data, cutoff, fs, order=5):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

## NOTCH FILTER CODE
def butter_notch(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='stop')
    return b, a


def butter_notch_filter(data, cutoff, fs, order=5): #cutoff set as 50 Hz
    b, a = butter_notch(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

# Example usage
fs = 1000.0  # Sampling frequency
highpass_cutoff = 1.0  # High-pass cutoff frequency
notch_cutoff = 50.0  # Notch filter cutoff frequency (for power line noise)
bandpass_lowcut = 8.0  # Lower cutoff frequency for band-pass filter
bandpass_highcut = 13.0  # Higher cutoff frequency for band-pass filter

df_cleaned_values = np.array(df_cleaned_values, dtype=float)  # Ensures correct type

# Apply high-pass filter
data_highpass = butter_highpass_filter(df_cleaned_values, highpass_cutoff, fs)

# Apply notch filter to remove 50 Hz power line noise
data_notch = butter_notch_filter(data_highpass, notch_cutoff, fs)

# Apply band-pass filter (8-13 Hz) for alpha band
filtered_data = butter_bandpass_filter(data_notch, bandpass_lowcut, bandpass_highcut, fs)

# Plot the results
plt.figure(figsize=(10, 6))
plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Original EEG Signal')

plt.subplot(4, 1, 2)
plt.plot(data_highpass)
plt.title('High-pass Filtered EEG Signal')

plt.subplot(4, 1, 3)
plt.plot(data_notch)
plt.title('Notch Filtered EEG Signal')

plt.subplot(4, 1, 4)
plt.plot(filtered_data)
plt.title('Band-pass Filtered EEG Signal')

plt.tight_layout()
plt.show()
```
