### 1: high pass -> notch filter
    - notch targets narrow frequency band VS high pass removes a broad range of low frequences
    **highpass**
    removes baseline drifts (slow  unwanted changes in signals baseline level) removes 0.5/1 Hz etc
    **notch fitler**
    - remove power line noise( 50 or 60 Hz) from  signal, creating a cleaner data set more suitable for extracting meaningful features related to emotional states
    - improves the accuracy of emotion classification algorithms.
    notch filtering = initial step
    
### 2: Band Pass filtering
    - filter to extract specific frequencies by iodentifying the low cutoff and high cutoff frequencis

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Filter functions
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order)
    return lfilter(b, a, data)

def butter_highpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

def butter_highpass_filter(data, cutoff, fs, order=5):
    b, a = butter_highpass(cutoff, fs, order)
    return lfilter(b, a, data)

def butter_notch(cutoff, fs, order=5):
    nyq = 0.5 * fs
    notch_band = [cutoff - 2, cutoff + 2]  # Define a small stopband around cutoff
    b, a = butter(order, [notch_band[0] / nyq, notch_band[1] / nyq], btype='bandstop')
    return b, a

def butter_notch_filter(data, cutoff, fs, order=5):
    b, a = butter_notch(cutoff, fs, order)
    return lfilter(b, a, data)

# Sampling frequency and cutoff frequencies
fs = 1000.0
highpass_cutoff = 1.0
notch_cutoff = 50.0
bandpass_lowcut = 8.0
bandpass_highcut = 13.0

# Convert EEG data to NumPy array
data = df_cleaned.iloc[:, 1:9]  # Select EEG channels
df_cleaned_values = np.array(data.values, dtype=float)

# Apply filters column-wise (each EEG channel separately)
data_highpass = np.apply_along_axis(butter_highpass_filter, 0, df_cleaned_values, highpass_cutoff, fs)
data_notch = np.apply_along_axis(butter_notch_filter, 0, data_highpass, notch_cutoff, fs)
filtered_data = np.apply_along_axis(butter_bandpass_filter, 0, data_notch, bandpass_lowcut, bandpass_highcut, fs)

# Plot results for the first EEG channel
plt.figure(figsize=(10, 6))
plt.subplot(4, 1, 1)
plt.plot(df_cleaned_values[:, 0])  # First channel
plt.title('Original EEG Signal')

plt.subplot(4, 1, 2)
plt.plot(data_highpass[:, 0])
plt.title('High-pass Filtered EEG Signal')

plt.subplot(4, 1, 3)
plt.plot(data_notch[:, 0])
plt.title('Notch Filtered EEG Signal')

plt.subplot(4, 1, 4)
plt.plot(filtered_data[:, 0])
plt.title('Band-pass Filtered EEG Signal')

plt.tight_layout()
plt.show()
```

```python
#Spectrogram

from matplotlib import pyplot as plt
from scipy import signal
import numpy as np

# Increase figure size for more space
plt.figure(figsize=(20, 20))

# Small constant to avoid log(0)
epsilon = 1e-10

# Loop through each EEG channel
for i in range(df_cleaned_values.shape[1]):
    # Compute the spectrogram for each channel
    f, t, Sxx = signal.spectrogram(df_cleaned_values[:, i], fs)

    # Apply log with a small epsilon to avoid log(0)
    Sxx_log = np.log(Sxx + epsilon)

    # Plot the spectrogram for each channel
    ax = plt.subplot(df_cleaned_values.shape[1], 1, i+1)
    #plt.subplot(df_cleaned_values.shape[1], 1, i+1)
    plt.pcolormesh(t, f, Sxx_log)
    plt.title(f'Spectrogram for Channel {i+1}')
    plt.xlabel('Time [sec]')
    ax.set_aspect('auto', adjustable='box')  # Ensure aspect ratio is adjusted


# Adjust the spacing between subplots
plt.subplots_adjust(hspace=.5, wspace=.1)  # Adjust vertical and horizontal spacing

plt.show()
```

```python
#this is for feature extraction for next week

import numpy as np
import scipy.signal as signal
from scipy.stats import skew, kurtosis

def compute_psd(data, fs):
    freqs, psd = signal.welch(data, fs, nperseg=fs)  # 1-second windows
    return freqs, psd

def bandpower(psd, freqs, band, fs):
    idx_band = np.where((freqs >= band[0]) & (freqs <= band[1]))[0]
    return np.trapz(psd[idx_band], freqs[idx_band])

def extract_features(eeg_data, fs):
    """Extracts features from each EEG channel"""
    features = []

    for channel in eeg_data.T:  # Iterate over channels (columns)
        freqs, psd = compute_psd(channel, fs)

        # Compute power in each frequency band
        delta = bandpower(psd, freqs, (0.5, 4), fs)
        theta = bandpower(psd, freqs, (4, 8), fs)
        alpha = bandpower(psd, freqs, (8, 13), fs)
        beta = bandpower(psd, freqs, (13, 30), fs)
        gamma = bandpower(psd, freqs, (30, 45), fs)

        # Statistical features
        mean_val = np.mean(channel)
        var_val = np.var(channel)
        skew_val = skew(channel)
        kurt_val = kurtosis(channel)

        # Hjorth parameters
        mobility = np.sqrt(np.var(np.diff(channel)) / var_val)
        complexity = np.sqrt(np.var(np.diff(np.diff(channel))) / np.var(np.diff(channel)))

        features.extend([delta, theta, alpha, beta, gamma, mean_val, var_val, skew_val, kurt_val, mobility, complexity])

    return np.array(features)

# Example: Extract features from EEG data
eeg_features = extract_features(filtered_data, fs)
print("Extracted Features Shape:", eeg_features.shape)
```
